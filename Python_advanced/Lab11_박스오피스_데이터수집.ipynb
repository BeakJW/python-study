### for문으로 여러 개의 페이지 불러오기

```python
import requests
from pandas import DataFrame
```

```python
url = "https://dapi.kakao.com/v3/search/book"

#쿼리스트링 요청 변수

#검색어는 필수
query = "파이썬"
page = 1
size = 50
```

```python
# 용도: 크롤링 시 봇 차단 우회, 브라우저처럼 요청
# url만 수정해서 사용
resultset=[]

with requests.Session() as session:
    # 세션 객체에서 웹브라우저 정보 (user agent) 주입
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",
        "Authorization": "KakaoAK %s" %key
    })

    for i in range(1,51):
        r = session.get(url,params={"query":query,"page":i,"size":size})

        if r.status_code != 200:
            msg = "[%d ERROR] %s 에러가 발생함" % (r.status_code, r.reason)

            print(msg)
            continue
        
        print("%d page 데이터 수신완료 >> %s"%(i,r))
        r.encoding = 'utf-8'
        mydict = r.json()
        resultset.extend(mydict['documents'])
print("총 %d건의 데이터가 수집됨"%len(resultset))
```

```python
df = DataFrame(resultset)
df.to_excel("카카오 책 검색(%s)-전체 데이터.xlsx"%query)
```

```python
def download(url,target):
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",
        "Authorization": "KakaoAK %s" %key
    })

    try:
        r = session.get(url,stream=True)
        r.encoding = "utf-8"

        with open(target,"wb") as f:
            f.write(r.raw.read())
            print(target,"가 저장되었습니다.")
    except Exception as e:
        print(target,"저장실패",e)
```

```python
import os
import datetime as dt

dirname = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
os.mkdir(dirname)

for i,v in enumerate(resultset):
    file_path = os.path.join(dirname,"%d.jpg"%i)

    download(v['thumbnail'],file_path)
```

```python
from concurrent import futures

dirname = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
os.mkdir(dirname)

with futures.ThreadPoolExecutor() as executor:

    for i,v in enumerate(resultset):
        file_path = os.path.join(dirname,"%d.jpg"%i)

        executor.submit(download,v['thumbnail'],file_path)
```

```python
data = {
    "startDate": startDate,    # 검색 시작일
    "endDate": today,          # 검색 종료일
    "timeUnit": "date",        # date: 일간, week: 주간, month: 월간
    "keywordGroups": [
        {
            "groupName": "Benz",  # 검색 주제
            "keywords": ["벤츠", "e클래스", "s클래스", "c클래스", "벤츠 cls"]  # 관련 검색어
        },
        {
            "groupName": "BMW",
            "keywords": ["BMW", "5시리즈", "x5", "x7", "x3"]
        },
        {
            "groupName": "AUDI",
            "keywords": ["아우디", "a6", "e트론", "a7", "q5"]
        }
    ]
}
```

```python
with requests.Session() as session:
    # 세션 객체에서 웹브라우저 정보 (user agent) 주입
    session.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",
        "X-Naver-Client-Id": clientId,
        "X-Naver-Client-Secret": clientSecret
    })
    r = session.post(url,data=json.dumps(data))

    if r.status_code != 200:
        msg = "[%d ERROR] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    print(r)
```

```python
r.encoding='utf-8'
mydict=r.json()
mydict["results"]
```

```python
resultset=[]

for item in mydict["results"]:
    title = item["title"]
    data= item["data"]

    for i,v in enumerate(data):
        new_data = {"title":title,"period":v["period"],"ratid":v["ratio"]}
        data[i] = new_data

    resultset.extend(data)

resultset
```
