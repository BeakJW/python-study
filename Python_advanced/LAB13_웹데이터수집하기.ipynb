```python
import requests
from bs4 import BeautifulSoup
```

```python
with requests.Session() as session:
    # 세션 객체에서 웹브라우저 정보 (user agent) 주입
    session.headers.update(
        {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"
        }
    )

    url = "https://data.hossam.kr/py/sample.html"
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d ERROR] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    print(r)
```

```python
r.encoding = "utf-8"

print(type(r.text))
r.text
```

## 연구과제

```python
import requests
from bs4 import BeautifulSoup
```

```python
with requests.Session() as session:
    # 세션 객체에서 웹브라우저 정보 (user agent) 주입
    session.headers.update(
        {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"
        }
    )

    url = "https://data.hossam.kr/py/myfood.html"
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d ERROR] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    print(r)
```

```python
r.encoding = "utf-8"

print(type(r.text))
r.text
```

```python
soup = BeautifulSoup(r.text)
print(type(soup))
soup
```

```python
img = soup.select(".img-wrapper img")
name = soup.select(".food-content h2")
desc = soup.select(".food-content p")

# img
# name
# desc

img_list = []
name_list = []
desc_list = []
```

```python
for i in img:
    img_list.append(i["src"])
for v in name:
    name_list.append(v.text)
for d in desc:
    desc_list.append(d.text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "name": name_list,
        "description": desc_list,
        "image": img_list,
    }
)

df.to_excel("food_data.xlsx", index=False)
```
